```python
#Training, Testing and Measuring Accuracy

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#Importing our Data
musicData=pd.read_csv("music.csv")
#Cleaning the Data/ Refining The Data
X=musicData.drop(columns=['genre'])
Y=musicData['genre']
#Splitting our data for training and testing
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.7)
#Create Model
model=DecisionTreeClassifier()
#Train the model so it can Learn Pattren
model.fit(x_train,y_train)

prediction=model.predict(x_test)
score=accuracy_score(y_test,prediction)
score
```

```python
#Training, Testing and Measuring Accuracy

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#Importing our Data
musicData=pd.read_csv("music.csv")
#Cleaning the Data/ Refining The Data
X=musicData.drop(columns=['genre'])
Y=musicData['genre']
#Splitting our data for training and testing
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.7)
#Create Model
model=DecisionTreeClassifier()
#Train the model so it can Learn Pattren
model.fit(x_train,y_train)
print("x_train Data -> ",x_train)
print("y_train Data -> ",y_train)
print("Prediction For -> ",x_test)
prediction=model.predict(x_test)
print("Prediction Result -> ",prediction)
print("y_test -> ",y_test)
score=accuracy_score(y_test,prediction)
print("Accuracy Score -> ",score)
```
